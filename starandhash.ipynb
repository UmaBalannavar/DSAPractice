You are given a string S consisting only of '*' and '#'.
You must find the difference between the number of '*' and the number of '#'.

If both are equal → print 0

If * are more → print positive difference

If # are more → print negative difference

```python
S = input().strip()

count_star = S.count('*')
count_hash = S.count('#')

result = count_star - count_hash
print(result)
```

Explain the code briefly.

input().strip() → reads input and removes extra spaces

.count('*') → counts how many stars

.count('#') → counts how many hashes

Subtract and print the result

time complexity?
➡️ O(n) — where n is the length of the string.
